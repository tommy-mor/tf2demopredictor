```python
import torch
import numpy as np
import torch.nn as nn
```

```python
import graph

data1 = graph.read_file('file.txt')
data2 = graph.read_file('file2.txt')
```

```python
import matplotlib.pyplot as plt

plt.plot(data1[0].sum(axis=0))
plt.plot(data1[1].sum(axis=0))
```

```python
# x = [myhealhts, enemyhealths], y = [mysuminfuture, enemysuminfuture]
X = []
Y = []
(_, _, max_tick) = data2.shape
i = 0
distance = 66 * 10
while i < max_tick - distance*2:
    i += 1
    myhealths = data2[0,:,i]
    enemyhealths = data2[1,:,i]
    mysuminfuture = data2[0,:,i+distance].sum()
    enemysuminfuture = data2[1,:,i+distance].sum()
    X.append(np.concatenate((myhealths, enemyhealths)))
    Y.append(np.array([mysuminfuture, enemysuminfuture]))

X = np.array(X)
Y = np.array(Y)
```

```python
training_x = torch.tensor(X, dtype=torch.float32)
training_y = torch.tensor(Y, dtype=torch.float32)

n_samples, n_features = training_x.shape
print(n_samples, n_features)

n_samples, n_output_features = training_y.shape
print(n_output_features)
```

```python
class Predictor(nn.Module):
    def __init__(self, input, output):
        super(Predictor, self).__init__()

        self.first = nn.Linear(input, 10)
        self.third = nn.Linear(input, 10)
    
    def forward(self, x):
        return self.second(self.first(x))
    
model = Predictor(n_features, n_output_features)
```

```python
model(training_x[0])
```

```python
learning_rate = 0.001
n_iters = 1000

losses = []

loss = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)

print(training_y[0])
for epoch in range(n_iters):
    y_pred = model(training_x)
    l = loss(training_y, y_pred)
    l.backward()

    optimizer.step()
    optimizer.zero_grad()

    if epoch % 10 == 0:
        with torch.no_grad():
            print(f'epoch {epoch} model{model(training_x[0])}loss ={l}')
            losses.append(float(l))
```

```python
plt.plot(losses)
```

```python
%matplotlib qt
with torch.no_grad():
    x = model(training_x)
plt.plot(training_y[:,1], color='blue')
plt.plot(x[:,1], color='red')
```
